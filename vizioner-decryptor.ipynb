```python
from math import gcd, sqrt
```

```python
def get_divisors(n):
    divisors = set()
    for i in range(1, int(sqrt(n)) + 1):
        if n % i == 0:
            divisors.add(i)
            divisors.add(n // i)
    return sorted(divisors)
```

```python

count_entities = 1
letter_len = 4
```

```python
trigs = set()
for i in range(0, len(crypto_text)-letter_len):
    if crypto_text.count(crypto_text[i:i+letter_len]) > count_entities:
        trigs.add(crypto_text[i:i+letter_len])
```

```python
period_lens = set()
for trig in list(trigs):
    first_index = crypto_text.find(trig)
    second_index = crypto_text.find(trig, first_index+letter_len)
    period = second_index - first_index - 1
    period_lens.add(period)
```

#### Key of len is in dels of 26 -- 100%

```python
for i in list(period_lens):
    for e in get_divisors(26):
        if gcd(i, e) > 2 and i != e:
            print(i, e, gcd(i, e))
```
